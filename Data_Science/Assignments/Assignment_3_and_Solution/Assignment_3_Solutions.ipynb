# Q1. Scenario: A company wants to analyze the sales performance of its products in different regions. They have collected the following data:<br>
   
Region A: [10, 15, 12, 8, 14]<br>
Region B: [18, 20, 16, 22, 25]
   
Calculate the mean sales for each region.

```python
region_a_sales = [10, 15, 12, 8, 14]
region_b_sales = [18, 20, 16, 22, 25]

mean_region_a = np.mean(region_a_sales)
mean_region_b = np.mean(region_b_sales)

print("Mean sales for Region A:", mean_region_a)
print("Mean sales for Region B:", mean_region_b)
```

```python
regions = ['Region A', 'Region B']
sales_data = [region_a_sales, region_b_sales]
mean_sales = [mean_region_a, mean_region_b]

# Plotting bar chart
plt.bar(regions, [np.mean(data) for data in sales_data], label='Mean Sales')
for i in range(len(sales_data)):
  plt.bar(regions[i], [data for data in sales_data[i] if data <= mean_sales[i]], alpha=0.6, label='Sales Data')

# Adding labels and title
plt.xlabel('Region')
plt.ylabel('Sales')
plt.title('Sales Performance by Region')

# Adding mean values as annotations
for i in range(len(regions)):
    plt.text(i, mean_sales[i]+1, f'{mean_sales[i]:.2f}', ha='center', color='black')

# Displaying legend and showing the plot
plt.legend()
plt.show()
```

# Observation :
The bar chart visualization provides a clear comparison of the sales performance in Region A and Region B.

1. Mean Sales: The bar labeled "Mean Sales" represents the average sales value for each region. It provides a reference point to compare the actual sales data.

2. Sales Data: The transparent bars represent the actual sales data for each region. By comparing the height of these bars, we can see the variations in sales between different data points.

3. Region A vs. Region B: From the plot, we can observe that Region B generally has higher sales compared to Region A. This can be seen from the higher position of the "Sales Data" bars for Region B.

4. Variability: The length of the "Sales Data" bars indicates the variability or range of sales values within each region. In both regions, the bars show some variation, but Region B appears to have a slightly wider range of sales.

5. Mean Annotations: The mean sales values are displayed as annotations on top of each bar. These values provide a precise representation of the average sales for each region. By comparing the mean sales with the sales data bars, we can quickly assess whether a particular data point is above or below the average.

# Q5. Scenario: A study is conducted to compare the performance of two different teaching methods. The test scores of the students in each group are as follows:
   
   Group A: [85, 90, 92, 88, 91]
  
   Group B: [82, 88, 90, 86, 87]
  
   Perform a t-test to determine if there is a significant difference in the mean scores between the two groups.


```python
group_a_scores = [85, 90, 92, 88, 91]
group_b_scores = [82, 88, 90, 86, 87]

t_statistic, p_value = stats.ttest_ind(group_a_scores, group_b_scores)

alpha = 0.05

print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

if p_value < alpha:
    print("There is a significant difference between the mean scores of the two groups.")
else:
    print("There is no significant difference between the mean scores of the two groups.")
```

The code performs a two-sample t-test to compare the mean scores between Group A and Group B.

1. T-Statistic: The t-statistic measures the difference between the mean scores of the two groups relative to the variation within each group. A larger absolute t-statistic indicates a larger difference between the means.

2. P-Value: The p-value represents the probability of observing the data (or more extreme data) assuming that the null hypothesis is true. It indicates the level of evidence against the null hypothesis. A smaller p-value suggests stronger evidence against the null hypothesis.

3. Significance Level: The significance level (commonly denoted as alpha) determines the threshold for determining statistical significance. If the p-value is less than alpha, we reject the null hypothesis and conclude that there is a significant difference between the mean scores of the two groups.

## Conclusion:
- Based on the comparison of the p-value and the significance level, we can make a conclusion about the significant difference between the mean scores of the two groups.

```python
group_a_scores = [85, 90, 92, 88, 91]
group_b_scores = [82, 88, 90, 86, 87]

groups = ['Group A', 'Group B']
scores = [group_a_scores, group_b_scores]

# Calculate the mean scores for each group
mean_scores = [np.mean(scores[i]) for i in range(len(scores))]

# Plotting bar chart
plt.bar(groups, mean_scores)
plt.xlabel('Group')
plt.ylabel('Mean Score')
plt.title('Mean Test Scores by Group')

# Adding mean scores as annotations
for i, score in enumerate(mean_scores):
    plt.text(groups[i], score + 0.5, f'{score:.2f}', ha='center')

plt.show()
```

# Q8. Scenario: A company wants to analyze the relationship between employee tenure and job satisfaction. The data collected is as follows:

   Employee Tenure (in years): [2, 3, 5, 4, 6, 2, 4]

   Job Satisfaction (on a scale of 1 to 10): [7, 8, 6, 9, 5, 7, 6]
   
   Perform a linear regression analysis to predict job satisfaction based on employee tenure.


```python
employee_tenure = [2, 3, 5, 4, 6, 2, 4]
job_satisfaction = [7, 8, 6, 9, 5, 7, 6]

# Reshape the data to fit the format expected by scikit-learn
X = np.array(employee_tenure).reshape(-1, 1)
y = np.array(job_satisfaction)

# Create a LinearRegression model and fit the data
model = LinearRegression()
model.fit(X, y)

# Make predictions on the training data
y_pred = model.predict(X)

# Predict job satisfaction for a given employee tenure
new_tenure = np.array([[1.2]])  # Example tenure value for prediction
predicted_satisfaction = model.predict(new_tenure)

# Calculate performance metrics
mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("R-squared (R2):", r2)

print("Predicted Job Satisfaction:", predicted_satisfaction[0])
```

```python
# Predict job satisfaction for the given employee tenure
new_tenure = np.linspace(2, 6, 100).reshape(-1, 1)  # Generate tenure values for plotting
predicted_satisfaction = model.predict(new_tenure)

# Plot the data points and the regression line
plt.scatter(X, y, label='Data')
plt.plot(new_tenure, predicted_satisfaction, color='red', label='Linear Regression')

plt.xlabel('Employee Tenure (in years)')
plt.ylabel('Job Satisfaction')
plt.title('Linear Regression: Employee Tenure vs. Job Satisfaction')

plt.legend()
plt.show()
```

## Overfitting:

- Overfitting occurs when the model fits the training data too closely, resulting in poor generalization to new data. To assess if overfitting has occurred, it's crucial to evaluate the model's performance on unseen data, such as a separate test dataset.

# Q9. Scenario: A study is conducted to compare the effectiveness of two different medications. The recovery times of the patients in each group are as follows:

   Medication A: [10, 12, 14, 11, 13]

   Medication B: [15, 17, 16, 14, 18]
   
   Perform an analysis of variance (ANOVA) to determine if there is a significant difference in the mean recovery times between the two medications.


```python


medication_a = [10, 12, 14, 11, 13]
medication_b = [15, 17, 16, 14, 18]

# Perform one-way ANOVA
f_value, p_value = stats.f_oneway(medication_a, medication_b)

print("F-value:", f_value)
print("p-value:", p_value)
```

## Observation:

- If the p-value is less than the chosen significance level (e.g., 0.05), we can reject the null hypothesis and conclude that there is a significant difference in the mean recovery times between the two medications.

- If the p-value is greater than the significance level, we fail to reject the null hypothesis and conclude that there is not enough evidence to suggest a significant difference in the mean recovery times.

- It's important to note that ANOVA assumes certain assumptions, such as normality and homogeneity of variances, which should be checked before drawing final conclusions. Additionally, post-hoc tests, such as Tukey's HSD or Bonferroni correction, can be performed to identify specific differences between groups if the ANOVA indicates a significant difference.

## Conclusion
- The obtained p-value of 0.0039, which is less than the significance level of 0.05, indicates that there is a significant difference in the mean recovery times between Medication A and Medication B. Therefore, we can conclude that the effectiveness of the two medications, as measured by the recovery times, is significantly different

# 13. Scenario: A survey is conducted to measure customer satisfaction with a product on a scale of 1 to 10. The data collected is as follows:

  [7, 9, 6, 8, 10, 7, 8, 9, 7, 8]
    
  Calculate the 95% confidence interval for the population mean satisfaction score.

```python
satisfaction_scores = [7, 9, 6, 8, 10, 7, 8, 9, 7, 8]

# Calculate sample mean and standard error
sample_mean = np.mean(satisfaction_scores)
standard_error = stats.sem(satisfaction_scores)

# Calculate confidence interval
confidence_interval = stats.t.interval(0.95, len(satisfaction_scores) - 1, loc=sample_mean, scale=standard_error)

print("95% Confidence Interval:", confidence_interval)
```

## Observation:

- The lower bound of the confidence interval is 7.0436, and the upper bound is 8.7564.
- This means that we can be 95% confident that the true population mean satisfaction score falls within this range.

## Conclusion:

- The obtained confidence interval suggests that the population mean satisfaction score is likely to be between 7.0436 and 8.7564, with a 95% confidence level.
- This implies that, based on the sample data collected, we can be reasonably confident that the true mean satisfaction score of the entire population lies within this interval.
- The range of the confidence interval provides an estimate of the uncertainty associated with the sample data and serves as a plausible range of values for the population mean.
- It is important to note that the confidence interval does not guarantee that the true population mean falls within this range with 100% certainty, but it provides a measure of the precision and reliability of our estimate based on the sample data.


Overall, based on the obtained confidence interval, we can conclude that there is a high likelihood that the true population mean satisfaction score lies within the range of 7.0436 to 8.7564, with a 95% confidence level.







# 14. Scenario: A company wants to analyze the effect of temperature on product performance. The data collected is as follows:

  Temperature (in degrees Celsius): [20, 22, 23, 19, 21]

  Performance (on a scale of 1 to 10): [8, 7, 9, 6, 8]
    
  Perform a simple linear regression to predict performance based on temperature.


```python
temperature = np.array([20, 22, 23, 19, 21]).reshape((-1, 1))
performance = np.array([8, 7, 9, 6, 8])

# Create and fit the linear regression model
model = LinearRegression()
model.fit(temperature, performance)

# Get the slope and intercept of the linear regression line
slope = model.coef_[0]
intercept = model.intercept_

# Predict performance for the given temperatures
predicted_performance = model.predict(temperature)

# Evaluate model performance
mse = mean_squared_error(performance, predicted_performance)
r2 = r2_score(performance, predicted_performance)

# Print model coefficients and performance measurements
print("Slope:", slope)
print("Intercept:", intercept)
print("Mean Squared Error (MSE):", mse)
print("R-squared (R2):", r2)
```

```python
# Plot the data points and the linear regression line
plt.scatter(temperature, performance, color='blue', label='Actual Performance')
plt.plot(temperature, predicted_performance, color='red', label='Predicted Performance')
plt.xlabel('Temperature')
plt.ylabel('Performance')
plt.title('Linear Regression: Performance vs Temperature')
plt.legend()
plt.show()
```

## Observations:

1. Slope: The obtained slope of 0.5 suggests that, on average, for every 1-unit increase in temperature, the predicted performance increases by 0.5 units. This positive slope indicates a positive linear relationship between temperature and performance.

2. Intercept: The intercept of -2.9 represents the estimated performance when the temperature is zero. However, in this scenario, since the temperature values are within a specific range, interpreting the intercept as a practical value might not be meaningful.

3. Mean Squared Error (MSE): The MSE of 0.54 indicates the average squared difference between the predicted performance values and the actual performance values. A lower MSE indicates better model fit. In this case, the MSE suggests a relatively low average squared difference, which implies that the model's predictions are reasonably close to the actual performance values.

4. R-squared (R2): The R2 value of 0.48 indicates that approximately 48% of the variance in the performance can be explained by the linear relationship with temperature. This value suggests a moderate level of goodness-of-fit, meaning that the model captures about 48% of the variability in performance based on temperature.

## Conclusion:

- The positive slope indicates that as the temperature increases, the predicted performance tends to increase as well. However, it is essential to note that this is a simplified interpretation based on the linear regression model and may not reflect the entire complexity of the relationship between temperature and performance.

- The obtained MSE value of 0.54 indicates that, on average, the predictions of the model deviate by approximately 0.54 units from the actual performance values. This suggests that the model's predictions are relatively close to the true performance values.

- The R2 value of 0.48 suggests that approximately 48% of the variance in the performance can be explained by the linear relationship with temperature. This indicates a moderate level of explanatory power, indicating that other factors beyond temperature may also influence performance.


Overall, based on the slope, MSE, and R2 values, it can be concluded that the linear regression model captures a moderate level of the relationship between temperature and performance.

# 15. Scenario: A study is conducted to compare the preferences of two groups of participants. The preferences are measured on a Likert scale from 1 to 5. The data collected is as follows:

  Group A: [4, 3, 5, 2, 4]

  Group B: [3, 2, 4, 3, 3]
    
  Perform a Mann-Whitney U test to determine if there is a significant difference in the median preferences between the two groups.

```python
group_a = np.array([4, 3, 5, 2, 4])
group_b = np.array([3, 2, 4, 3, 3])

# Perform Mann-Whitney U test
statistic, p_value = mannwhitneyu(group_a, group_b)

# Print test statistic and p-value
print("Mann-Whitney U statistic:", statistic)
print("p-value:", p_value)
```

## Observations:

1. Mann-Whitney U statistic: The obtained Mann-Whitney U statistic is 17.0. This statistic represents the sum of ranks for the observations in Group A. It measures the difference in ranks between the two groups and is used to test the hypothesis of no difference in median preferences.

2. p-value: The p-value is calculated as 0.380836480306712. It indicates the probability of obtaining the observed difference in ranks between the two groups (or a more extreme difference) under the assumption of no difference in median preferences. A higher p-value suggests weaker evidence against the null hypothesis.

## Conclusion:

Based on the obtained results:

1. The Mann-Whitney U statistic of 17.0 indicates the observed difference in ranks between Group A and Group B.
2. The p-value of 0.380836480306712 is greater than the commonly used significance level of 0.05. This suggests that there is insufficient evidence to reject the null hypothesis of no difference in median preferences between the two groups.
3. Therefore, we do not have enough evidence to conclude that there is a significant difference in the median preferences between Group A and Group B. The data does not provide strong support for the presence of a systematic difference in preferences between the two groups.


It is important to note that the Mann-Whitney U test is a nonparametric test that compares the distributions of two independent groups. It does not make assumptions about the underlying distribution of the data and is suitable when the assumptions of parametric tests (such as t-test) are not met. In this case, the obtained p-value suggests that there is no significant difference in the median preferences between the two groups based on the Likert scale data

  Algorithm C: [0.90, 0.88, 0.89, 0.86, 0.87]
    
  Perform a Kruskal-Wallis test to determine if there is a significant difference in the median accuracy scores between the algorithms.

```python
algorithm_A = [0.85, 0.80, 0.82, 0.87, 0.83]
algorithm_B = [0.78, 0.82, 0.84, 0.80, 0.79]
algorithm_C = [0.90, 0.88, 0.89, 0.86, 0.87]

# Perform the Kruskal-Wallis test
h_statistic, p_value = kruskal(algorithm_A, algorithm_B, algorithm_C)

# Print the test results
print("Kruskal-Wallis H-statistic:", h_statistic)
print("p-value:", p_value)
```

## Observations:

1. The Kruskal-Wallis H-statistic value of 9.696947935368053 represents the test statistic that measures the difference in ranks between the accuracy scores of the three algorithms.
2. The obtained p-value of 0.007840333026249539 is below the significance level of 0.05 (assuming a commonly used significance level). This indicates that there is a statistically significant difference in the median accuracy scores between the algorithms.

## Conclusion:

1. The Kruskal-Wallis test suggests that there is a significant difference in the median accuracy scores among the three algorithms.
2. The obtained p-value of 0.007840333026249539 provides strong evidence against the null hypothesis of no difference in median accuracy scores.
3. Therefore, we can conclude that there is a statistically significant difference in the performance of the algorithms based on the accuracy scores.
